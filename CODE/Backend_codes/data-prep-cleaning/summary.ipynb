```python
cleandata = pandas.read_csv('/data/TransactDataClean_3110.csv')
cleandata.head()
```

```python
cleandata['InvoiceDate'] = cleandata['InvoiceDate'].astype('datetime64[ns]')
cleandata['Year'] = cleandata['InvoiceDate'].astype('datetime64[ns]').dt.year
cleandata = cleandata[cleandata['Quantity'] > 0]
# cleandata['Quantity'] = cleandata['Quantity'].abs()
cleandata.head()
cleandata.shape
```

```python
cleandata[cleandata.Cost < 0 ]
```

```python
cleandata['Cost'] =  cleandata['Quantity'] * cleandata['Price']
```

```python
temp=cleandata.groupby(["Country","Year","Customer ID"], as_index=False)["Cost"].agg(["sum"]).rename(columns = {'sum' :'sales' })

temp = temp.groupby(["Country","Year"], as_index=False)["sales"].agg(["sum","mean"]).rename(columns = {'sum':'Total Sales','mean':'Average Sales per Customer'})
print(temp)
```

```python
temp2= cleandata.groupby(["Country","Year",], as_index=False).agg({ "Customer ID": lambda x: x.nunique() }).rename(columns = {'Customer ID':'Customer Concentration'})
temp2
```

```python
# cleandata.groupby(["Country","Year"], as_index=False).agg( {'Customer ID' :pd.Series.unique })
```

```python
temp3 = pd.merge(temp,temp2,on=['Country','Year'])
```

```python
temp4=cleandata.groupby(["Country","Year","Invoice"], as_index=False)["Cost"].agg(["sum"]).rename(columns = {'sum' :'sales' })

temp4 = temp4.groupby(["Country","Year"], as_index=False)["sales"].agg(["mean"]).rename(columns = {'mean':'Average Sales per Transaction'})
print(temp4)
```

```python
temp5 = pd.merge(temp3,temp4,on=['Country','Year'])
temp5.info()
```

```python
temp6=cleandata.groupby(["Country","Year"], as_index=False)["Price"].agg(["mean"]).rename(columns = {'mean' :'Unit Price' })
```

```python
temp7 = pd.merge(temp5,temp6,on=['Country','Year'])
temp7.head()
```
